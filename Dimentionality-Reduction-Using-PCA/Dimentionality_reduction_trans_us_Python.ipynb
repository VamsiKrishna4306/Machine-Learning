# Dimensionality Reduction

###  Importing Required packages

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

### Importing dataset

```python
existing_df = pd.read_csv("D:/Bdat_Course_Material/Semister 2/Business Intelligence/Week 3/trans_us.csv",index_col=0,thousands=',')
existing_df.head()
```

```python
existing_df.index.names=['stations']
existing_df.columns.names=['months']
existing_df.head()
```

```python
existing_df = existing_df.fillna(15)
existing_df.head()
```

```python
from sklearn.decomposition import PCA
pca=PCA(n_components=2)
pca.fit(existing_df)
```

```python
existing_2d = pca.transform(existing_df)
existing_df_2d = pd.DataFrame(existing_2d)
existing_df_2d.columns = ['PC1','PC2']
existing_df_2d.head()
```

```python
print(pca.explained_variance_ratio_)
```
